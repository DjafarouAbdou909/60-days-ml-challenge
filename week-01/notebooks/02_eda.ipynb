```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv("../data/raw/Train.csv")
df_demo = pd.read_parquet("../data/raw/demographics_clean.parquet")
df_raw = df_train.merge(df_demo, on="UniqueID", how="left")
```

```python
df_raw["Age"] = (pd.Timestamp.today() - pd.to_datetime(df_raw["BirthDate"])).dt.days // 365
```

```python
df_raw["next_3m_txn_count"].hist()
df_raw["next_3m_txn_count"].describe()
```

```python
df_raw.groupby("IncomeCategory")["next_3m_txn_count"].mean().sort_values()
df_raw.groupby("Gender")["next_3m_txn_count"].mean()
df_raw.groupby("CustomerStatus")["next_3m_txn_count"].mean()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df_raw["next_3m_txn_count"], bins=50, color="steelblue")
axes[0].set_title("Distribution de next_3m_txn_count")
axes[0].set_xlabel("Transactions")

axes[1].hist(np.log1p(df_raw["next_3m_txn_count"]), bins=50, color="coral")
axes[1].set_title("Distribution log1p(next_3m_txn_count)")
axes[1].set_xlabel("log(Transactions)")

plt.tight_layout()
plt.savefig("../outputs/01_target_distribution.png")
plt.show()
```

```python
print(df_raw["next_3m_txn_count"].describe())
print(f"Skewness : {df_raw['next_3m_txn_count'].skew():.2f}")
```

```python
num_cols = ["Age", "AnnualGrossIncome", "next_3m_txn_count"]
corr = df_raw[num_cols].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Corrélations variables numériques")
plt.tight_layout()
plt.savefig("../outputs/02_correlations.png")
plt.show()
```

```python
cat_cols = [
    "Gender", "IncomeCategory", "CustomerStatus", "ClientType",
    "MaritalStatus", "OccupationCategory", "IndustryCategory",
    "CustomerBankingType", "CustomerOnboardingChannel",
    "LowIncomeFlag", "ContactPreference"
]

for col in cat_cols:
    plt.figure(figsize=(10, 4))
    order = df_raw.groupby(col)["next_3m_txn_count"].mean().sort_values(ascending=False).index
    sns.barplot(data=df_raw, x=col, y="next_3m_txn_count", order=order, palette="Blues_d")
    plt.title(f"Moyenne next_3m_txn_count par {col}")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(f"../outputs/03_cat_{col}.png")
    plt.show()
```

```python
print("\nQuantiles AnnualGrossIncome :")
print(df_raw["AnnualGrossIncome"].quantile([0.25, 0.50, 0.75, 0.95, 0.99, 0.999]))
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_raw["AnnualGrossIncome"], color="lightgreen")
plt.title("Boxplot AnnualGrossIncome")
plt.tight_layout()
plt.savefig("../outputs/04_outliers_income.png")
plt.show()
```

```python
print("\nQuantiles Age :")
print(df_raw["Age"].quantile([0.01, 0.25, 0.50, 0.75, 0.99]))
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df_raw["Age"], bins=40, color="mediumpurple")
plt.title("Distribution Age")
plt.tight_layout()
plt.savefig("../outputs/05_age_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df_raw["Age"], df_raw["next_3m_txn_count"], alpha=0.3, color="steelblue")
plt.xlabel("Age")
plt.ylabel("next_3m_txn_count")
plt.title("Age vs Transactions futures")
plt.tight_layout()
plt.savefig("../outputs/06_age_vs_target.png")
plt.show()
```
